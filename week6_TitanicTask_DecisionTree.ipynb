```python
import numpy as np
import sklearn as skl
import pandas as pd
```

# Read file train.csv

```python
train = pd.read_csv("train.csv")
train
```

```python
train.info()
```

```python
train.shape
```

```python
train_inputs = train.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis='columns')
train_target = train['Survived']
```

```python
train_target
```

```python
train_inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
sex_string = LabelEncoder()
```

```python
train_inputs['sex_to_number'] = sex_string.fit_transform(train_inputs['Sex'])
train_inputs.head(7)
```

```python
train_inputs_n = train_inputs.drop(['Sex'], axis='columns')
train_inputs_n
```

```python
train_inputs_n.info()
```

77 nan boxes in 'Age' column
=> replace nan with average number of column 'Age'

```python
for i in range(len(train_inputs_n.index)):
    print(train_inputs_n['Age'][i])
```

```python
avg = np.nanmean(train_inputs_n['Age'])
for i in range(len(train_inputs_n.index)):
    if np.isnan(train_inputs_n['Age'][i]):
        train_inputs_n.iloc[i, train_inputs_n.columns.get_loc('Age')] = avg
```

```python
for i in range(len(train_inputs_n.index)):
    print(train_inputs_n['Age'][i])
```

```python
train_inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(train_inputs_n, train_target)
```

```python
model.score(train_inputs_n, train_target)
```
